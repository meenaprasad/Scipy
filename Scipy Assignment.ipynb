```python
## calling Libraries & reading the required data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
## 1.Create a pie chart presenting the male/female proportion

titanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
titanic['sex'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, startangle=140)
plt.gca().set_aspect("equal")
plt.title(" Titanic - Male/Female proportion")
plt.show()
```

```python
## 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

sns.set_style("whitegrid");
sns.FacetGrid(titanic, hue="sex", height=6).map(plt.scatter, "fare", "age").add_legend()
plt.title("Titanic scatterplot with the Fare paid and the Age")
```
